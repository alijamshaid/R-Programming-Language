```r
veh=read.csv("E:/Results/Islamabad vehicles/TLZ_756.csv")#reading data from csv
veh[1:5,2:4] #specific column and rows from csv
head(veh)#first few rows
tail(veh)#last few rows
```

```r
hist(veh$fc_ml) #histogram
```

```r
hist(veh$fc_ml,
    breaks=10,
    xlim=c(0,10),
    ylim=c(0,1500),
    col='blue',
    xlab='Fuel Consumed',
    ylab='Fuel Consumption',
    main='Fuekl Consumption Histogram') #this shows colorfull histogram
```

```r
plot(veh$fc_ml,col='blue')#bar plot
```

```r
colors=c('green','red','yellow')
pie(table(veh$activity_type),
   col=colors)#how to draw pie chart
```
